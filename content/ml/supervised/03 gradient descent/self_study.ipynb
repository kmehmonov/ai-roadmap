```python
import numpy as np

def f(x):
    return x[0]**2 + x[1]**2 + 3*x[0]*x[1]

def numerical_gradient(f, x, h=1e-5):
    grad = np.zeros_like(x)
    for i in range(len(x)):
        x_step1 = x.copy()
        x_step2 = x.copy()
        x_step1[i] += h
        x_step2[i] -= h
        grad[i] = (f(x_step1) - f(x_step2)) / (2*h)
    return grad

x_point = np.array([1.0, 2.0])
grad = numerical_gradient(f, x_point)
print("Numerical gradient:", grad)
```

```python
%matplotlib widget

import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate grid data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
def f(x):
    return x[0]**2 + x[1]**2 + 3*x[0]*x[1]

Z= f(np.array([X, Y]))

# Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')  # 'plasma', 'coolwarm', etc.

# Labels
ax.set_title("3D Surface Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
